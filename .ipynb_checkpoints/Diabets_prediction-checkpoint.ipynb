## مقدمة عن مرض السكري

مرض السكري هو اضطراب مزمن يحدث عندما يصبح مستوى الجلوكوز (السكر) في الدم مرتفعًا عن المعدل الطبيعي. يحدث ذلك عادةً عندما لا ينتج الجسم كمية كافية من هرمون الإنسولين أو عندما لا يستطيع استخدامه بشكل فعال. الإنسولين هو هرمون يفرزه البنكرياس ويساعد الخلايا على امتصاص الجلوكوز من الدم لاستخدامه كمصدر للطاقة.

---

## أنواع مرض السكري

ينقسم مرض السكري إلى عدة أنواع رئيسية:

* **السكري من النوع الأول (Type 1 Diabetes):** يحدث عندما يهاجم الجهاز المناعي خلايا البنكرياس المسؤولة عن إنتاج الإنسولين، مما يؤدي إلى نقصه بشكل كبير. غالبًا ما يظهر هذا النوع في سن مبكرة.

* **السكري من النوع الثاني (Type 2 Diabetes):** وهو النوع الأكثر شيوعًا، ويحدث عندما يصبح الجسم مقاومًا لتأثير الإنسولين أو عندما لا ينتج البنكرياس كمية كافية منه. يرتبط هذا النوع غالبًا بعوامل نمط الحياة مثل السمنة وقلة النشاط البدني.

* **مرحلة ما قبل السكري (Prediabetes):** هي حالة تكون فيها مستويات السكر في الدم أعلى من الطبيعي ولكنها ليست مرتفعة بدرجة كافية لتشخيص مرض السكري.

---

## العوامل التي تؤثر على خطر الإصابة بالسكري

هناك مجموعة من العوامل التي قد تزيد من خطر الإصابة بمرض السكري، منها:

* **العمر:** يزداد خطر الإصابة بالسكري مع التقدم في العمر.
* **السمنة وزيادة الوزن:** خاصة تراكم الدهون في منطقة البطن.
* **قلة النشاط البدني:** النشاط البدني يساعد الجسم على استخدام الجلوكوز بكفاءة أكبر.
* **النظام الغذائي غير الصحي:** مثل الإفراط في تناول السكريات والكربوهيدرات المكررة.
* **التاريخ العائلي:** وجود حالات سكري في العائلة يزيد من احتمالية الإصابة.
* **التدخين:** قد يزيد من مقاومة الإنسولين ويؤثر سلبًا على الصحة العامة.

---

## المؤشرات الصحية المرتبطة بالسكري والقيم الطبيعية

لفهم بيانات السكري وتحليلها بشكل صحيح، من المهم معرفة المعدلات الطبيعية لبعض المؤشرات الصحية الشائعة.

### مستوى السكر في الدم (Blood Glucose)

| القياس                     | المعدل الطبيعي           |
| -------------------------- | ------------------------ |
| سكر صائم (Fasting Glucose) | من **70 إلى 99 mg/dL**   |
| مرحلة ما قبل السكري        | من **100 إلى 125 mg/dL** |
| سكري                       | **126 mg/dL أو أكثر**    |

---

### السكر التراكمي (HbA1c)

يقيس متوسط مستوى السكر في الدم خلال آخر 2–3 أشهر.

| الحالة              | القيمة               |
| ------------------- | -------------------- |
| طبيعي               | أقل من **5.7%**      |
| مرحلة ما قبل السكري | من **5.7% إلى 6.4%** |
| سكري                | **6.5% أو أكثر**     |

---

### مؤشر كتلة الجسم (BMI)

يستخدم لتحديد ما إذا كان الوزن طبيعيًا أو يشير إلى زيادة الوزن أو السمنة.

| الحالة    | BMI             |
| --------- | --------------- |
| نقص وزن   | أقل من **18.5** |
| وزن طبيعي | **18.5 – 24.9** |
| زيادة وزن | **25 – 29.9**   |
| سمنة      | **30 أو أكثر**  |

---

### ضغط الدم (Blood Pressure)

| الحالة       | الضغط                    |
| ------------ | ------------------------ |
| طبيعي        | أقل من **120 / 80 mmHg** |
| مرتفع قليلاً | **120–129 / أقل من 80**  |
| ضغط مرتفع    | **130 / 80 أو أكثر**     |

---

### الكوليسترول

| النوع             | المعدل الصحي                                      |
| ----------------- | ------------------------------------------------- |
| الكوليسترول الكلي | أقل من **200 mg/dL**                              |
| LDL (الضار)       | أقل من **100 mg/dL**                              |
| HDL (المفيد)      | أكثر من **40 mg/dL للرجال** و **50 mg/dL للنساء** |

---

```python
mask = data['gender'] == 'Other'

data.loc[mask, 'gender'] = np.random.choice(
    ['Male', 'Female'],
    size=mask.sum(),
    p=[0.5, 0.5]
)
data['gender'] = data['gender'].map({'Female':0, 'Male':1})
```

```python
data['ethnicity'].value_counts()
```

```python
mask = data['ethnicity'] == 'Other'

data.loc[mask, 'ethnicity'] = np.random.choice(
    ['White', 'Hispanic', 'Black', 'Asian'],
    size=mask.sum(),
    p=[44997, 20103, 17986, 11865] / np.sum([44997, 20103, 17986, 11865])
)
```

```python
data = pd.get_dummies(data, columns=['ethnicity'], drop_first=True)
```

```python
data['education_level'].value_counts()
```

```python
data = pd.get_dummies(data, columns=['education_level'], drop_first=True)
```

```python
income_mapping = {
    'Low': 0,
    'Lower-Middle': 1,
    'Middle': 2,
    'Upper-Middle': 3,
    'High': 4
}

data['income_level'] = data['income_level'].map(income_mapping)
```
